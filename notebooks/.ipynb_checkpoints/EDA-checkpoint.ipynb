```python
import pandas as pd

df = pd.read_csv("../data/train.csv")
df.head()
```

```python
df["is_toxic"] = (
    df[['toxic','severe_toxic','obscene','threat','insult','identity_hate']]
    .sum(axis=1) > 0
).astype(int)

df["is_toxic"].value_counts()
```

```python
df["is_toxic"].value_counts(normalize=True)
```

```python
import matplotlib.pyplot as plt

df["is_toxic"].value_counts().plot(kind="bar")
plt.title("Class Distribution (Toxic vs Non-Toxic)")
plt.show()
```

```python
df["text_length"] = df["comment_text"].str.len()
df["text_length"].hist(bins=50)
plt.title("Comment Length Distribution")
plt.show()
```
